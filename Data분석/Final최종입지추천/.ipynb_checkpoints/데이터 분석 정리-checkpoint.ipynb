이동형 수소충전시스템 
Mobile Hydrogen Charging System

**2018년도 한국에너지학회 춘계학술발표회**
- 화석연료 에너지 고갈 및 저유가시대 종언에 따라 기존 에너지자원은 이미 한계에 도달
- 교토의정서 발효 등으로 인해 전 세계적으로 온실가스 배출규제가 강화되고 있어 신·재생 에너지 요구가 가속화
- 수소에너지는 재생 가능한 무한자원으로 탄소배출이 없는 환경친화적 에너지로서, 향후 주요 에너지원이 될 것으로 전망됨
- 환경과 에너지문제를 일거에 해소


**2019.26.2.53-하이브리드형 수소공급방식을 고려한 수소충전소 입지 선정 모형 개발**
- 수소차 시장을 활성화하기 위해서는 수소충전 인프라 구축이 시급
- 효율적인 수소충전 인프라 구축을 위해서는 이용자와 공급자 측면에서의 다양한 요소를 고려한 계획적인 투자

- 수소충전 인프라 구축 초기단계에 적합한 하이브리드형 수소 공급방식 제안
- 공급방식 하에서 이용자 측명의 이용수요를 최대화하며 공급자 측면의 충전소/개질기 비용 및 수소운송비용을 최소화하는 수소 충전소 최적 입지 선정 모형 개발 

**HYGEN**
- 위와 같은 논문을 바탕으로 서울시를 타겟한 개질기와 이동식 수소충전소 최적 입지 선정 모형 개발  
- 수소의 새로운 시대를 여는 수소 충전소 종합 시스템

**Regional Consumer Hydrogen Demand and Optimal Hydrogen Refueling Station Siting**
(from the FY 2004 and 2005 work described in the Introduction above)

- 변수를 선택하는데 있어서 참고한 자료

**미국 Regional Consumer Hydrogen Demand and Optimal Hydrogen Refueling Station Siting 논문에 의하면 가구소득수준, 가구수별 승용차수, 미세먼지 정도, 출퇴근 거리, 교육, 수소차등록대수 등과 같은 변수들이 수소차를 사는 고객을 나타낼 수 있는 유의미한 Attributes**

**HYGEN 입지선정모델에 사용된 변수**
동별_대학교 이상 교육 받은 인구수 : 
    서울시 교육정도별 인구 - 서울 열린데이터 광장
동별_친환경자동차수(하이브리드, 전기차, 수소차) : 
    동별 자동차 등록 비율 * 구별 자동차 등록현황 - 서울 열린데이터 광장 
동별_사업체수 :
    산업별, 동별 사업체수 및 종사자수 - KOSIS국가통계포털
동별_㎡당 매매평균가 :
    가구소득과 유사
    KB주택가격동향 - KB 부동산
동별_가구수별_승용차수 :
    동별 자동차 등록현황 / 동별 가구수 - 서울 열린데이터 광장
동별_수소차_대수 :
    구별 수소차 대수 * 동별 자동차 등록 비율
동별_교통량 :
    서울시 택시운행 분석데이터 정보 - 서울 열린데이터 광장

* 7개의 변수 사용

```python
geo_path = "geoData_copy.json"
geo_str = json.load(open(geo_path, encoding="utf-8"))
```

```python
X.columns
```

```python
seoul_map = folium.Map(location=[37.5502, 126.982], zoom_start=10.5, tiles="cartodbpositron")
seoul_map.choropleth(geo_data=geo_str, 
                     data = X ,
                     columns=['동', '수소충전소'],
                     fill_color='YlGnBu',  # GnBu
                     key_on='properties.name',
                     highlight=True,
                     fill_opacity=0.5,
                     line_opacity=1,
                     legend_name="수소충전소")
seoul_map.save("수소충전소.html")
```

```python
# 양재 수소충전소를 찍어보자.
folium.Marker(
  location=[37.4686275,127.0319203],
  popup='양재 수소 충전소 위치.',
  icon=folium.Icon(color='red',icon='star')
).add_to(seoul_map)

# 상암 수소 충전소를 찍어보자.
folium.Marker(
  location=[37.5687204,126.8761933],
  popup='상암 수소 충전소 위치.',
  icon=folium.Icon(color='red',icon='star')
).add_to(seoul_map)

# 국회 수소 충전소를 찍어보자.
folium.Marker(
  location=[37.5311111,126.9132037],
  popup='2019년 완공 예정 국회 수소 충전소 위치.',
  icon=folium.Icon(color='blue',icon='star')
).add_to(seoul_map)

# 강동구 상일충전소를 찍어보자.
folium.Marker(
  location=[37.5392486,127.128611],
  popup='2019년 완공 예정 강동구 상일충전소 위치',
  icon=folium.Icon(color='blue',icon='star')
).add_to(seoul_map)
```
