```python
hydro = pd.read_csv("수소 위치.csv", encoding='utf-8')
```

```python
hydro=hydro[[       '지역',         '상호',        '운영시간',
             '주말운영', '충전비용(\/kg)',       '결제방식',            0,
                  1,            2,            3]]
```

```python
hydro.columns = [  '지역',         '상호',        '운영시간',
             '주말운영', '충전비용(\/kg)',       '결제방식',      '도시',
                  '시',            '구',            '동']
```

```python
columnsTitles = ['지역',  '도시',  '시',     '구',            '동'    ,   '상호',        '운영시간',
             '주말운영', '충전비용(\/kg)',       '결제방식'   ]
hydro= hydro.reindex(columns=columnsTitles)
```
